```python
import pandas as pd
```

```python
df = pd.read_csv("car_price.csv")
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
```

```python
dummies
```

```python
final_df = pd.concat([df,dummies], axis="columns")
```

```python
final_df
```

```python
final_df = final_df.drop(["Car Model", "Mercedez Benz C class"], axis="columns")
```

```python
final_df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X = final_df[["Mileage", "Age", "Audi A5", "BMW X5"]]
X
```

```python
y = final_df[["Sell Price"]]
y
```

```python
lr.fit(X,y)
```

```python
lr.predict([[69000,6,0,1]])
```
